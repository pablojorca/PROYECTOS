### Función que escribe las cartas

```python
def escriturafor(n):
    tpl=DocxTemplate("plantilla.docx") #cargar plantilla
    df=pd.read_excel("Datos_Proyectos.xlsx") #carga datos de excel
    x=df.to_dict(orient="records") #Se eligen las variables
    context=x
    df=pd.read_excel("Datos_Proyectos.xlsx")
    
    identificador=df["Codigo"].to_list() # Se crea el id de cada archivo
    
    #Rellenado 
    tpl.render(context[n])
    tpl.save("CartasAceptación\Carta de Aprobación Informe de Avance_%s.docx" %str(identificador[n]))
    wait=time.sleep(random.randint(0,1))
    
```

### Se ejecuta la función

```python
df2=len(df)
print("Habrán: ", df2, " archivos")

for i in range(0,df2):
    print("Escribiendo archivo:",f"{i+1},", "--Espere--")
    escriturafor(i)
```
