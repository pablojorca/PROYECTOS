### Se inicia sesión en la cuenta y luego

```python
smtp=smtplib.SMTP('smtp.gmail.com',587)
smtp.ehlo()
smtp.starttls()
smtp.login(email,password)


for i in range(0,len(df)):
    
    msgRoot = MIMEMultipart('related')
    msgRoot['Subject'] = "ENVÍO DE INFORMACIÓN PROYECTO CÓDIGO: "+str(codigo[i])
    
    msgRoot['From'] = email
    
    para=[lista[i],beneficiario[i]]
    
    msgRoot['To'] =', '.join(para)
    
    listo=','.join(para)
    
    cc=["pablo.jorquera@corfo.cl","consuelo.monje@corfo.cl","mario.palma@corfo.cl","ammerino@uc.cl"]
    
    msgRoot['Cc']=','.join(cc)
    
    para.extend(cc)
    
    msgRoot.preamble = '====================================================='
    msgAlternative = MIMEMultipart('alternative')
    msgRoot.attach(msgAlternative)
    msgText = MIMEText('This object contains the plain text content of this email.')
    # Attach the MIMEText object to the msgAlternative object.
    msgAlternative.attach(msgText)
    msgText = MIMEText(body, 'html')
    msgAlternative.attach(msgText)
    
    filename ="Comité\Planilla_Integrantes.xlsx"

    part = MIMEBase('application', "octet-stream")
    part.set_payload(open("Comité\Planilla_Integrantes.xlsx", "rb").read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', 'attachment; filename="Planilla_Integrantes.xlsx"')
    msgRoot.attach(part)
    
    fp = open('logo.png', 'rb')
    msgImage = MIMEImage(fp.read())
    fp.close()
    msgImage.add_header('Content-ID', '<image1>')
    msgRoot.attach(msgImage)
    print("Escribiendo para: "+ str(codigo[i]))
    smtp.sendmail(email, para, msgRoot.as_string())
    wait=time.sleep(random.randint(0,7))
smtp.quit()
```
